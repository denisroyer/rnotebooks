# BI - R-Kenntnis Übungen 00: Der Markdown Syntax

Wenn wir Texte lesen, ganz egal ob im Internet, in einer Zeitschrift oder einem gedruckten Buch, erwarten wir ein bestimmtes Format. Besonders wichtige Wörter sind gefettet, eine Überschrift lässt sich mit einem Blick vom restlichen Text unterscheiden und mit einer ordentlichen Liste wird das Geschriebene anschaulicher. Wir nehmen solche Formatierungen als selbstverständlich wahr, und wenn wir selbst einen Text am PC verfassen, gelingt uns das meist ohne Probleme: Schriftgröße anpassen, Spiegelstriche setzen und Wörter fetten. Jedes Textverarbeitungsprogramm bietet Nutzern die unterschiedlichsten Möglichkeiten, um ihr Geschriebenes ansprechend zu gestalten.

## Was ist Markdown?

Markdown ist eine vereinfachte Auszeichnungssprache, die von John Gruber und Aaron Swartz entworfen und im Dezember 2004 mit Version 1.0.1 spezifiziert wurde. Ein Ziel von Markdown ist eine leicht lesbare Ausgangsform bereits vor der Konvertierung. Als Auszeichnungselemente wurden daher vor allem Auszeichnungsarten verwendet, die in Plain text und E-Mails üblich sind.

## Wo wird Markdown genutzt?

Genutzt wird Markdown oder eine Markdown-ähnliche Syntax vorwiegend auf Entwicklerplattformen mit eher technikaffinem Publikum wie GitHub, Stack Overflow oder der Blogging-Plattform Ghost. Markdown wird häufig bei Readme-Dateien verwendet. Es kommt auch in populären webbasierten Tools wie der Projektmanagementsoftware Trello oder dem Instant-Messaging-Dienst Slack zum Einsatz.

# Der Markdown Syntax

Für die Auszeichnung (also die Formatierung) von Text verwendet Markdown vor allem Satzzeichen, um eine einfache Lesbarkeit zu ermöglichen:

## Normaler Text

<div class="alert alert-block alert-success">

**Ausgangsform:**

```Markdown
Normaler Text wird so dargestellt wie eingegeben.

Eine Leerzeile erzeugt einen Absatz.
```
</div>

**Zielform:**

<div class="alert alert-block alert-warning">
    
Normaler Text wird so dargestellt wie eingegeben.

Eine Leerzeile erzeugt einen Absatz.

</div>

**Hinweis:** Zwei oder mehr Leerzeichen am Ende der Zeile erzeugen einen Zeilenumbruch.

## Kursiv und Fett

<div class="alert alert-block alert-success">

**Ausgangsform:**

```Markdown
*Kursiv*, **Fett** und ***Fett kursiv*** bzw.

_Kursiv_, __Fett__ und ___Fett kursiv___
```
</div>

**Zielform:**

<div class="alert alert-block alert-warning">
    
*Kursiv*, **Fett** und ***Fett kursiv*** bzw.  
_Kursiv_, __Fett__ und ___Fett kursiv___

</div> 

## Inline-Quellentext

<div class="alert alert-block alert-success">

**Ausgangsform:**

```Markdown
Markiert Text als `Inline-Quelltext`
```
</div>

**Zielform:**

<div class="alert alert-block alert-warning">
    
Markiert Text als `Inline-Quelltext`

</div>

## Listen

### Bullet-Listen

<div class="alert alert-block alert-success">

**Ausgangsform:**

```Markdown
* Ein Punkt in einer ungeordneten Liste
* Ein weiterer Punkt in einer ungeordneten Liste
    * Ein Unterpunkt, um vier Leerzeichen eingerückt
* Statt * funktionieren auch + oder -
```
</div>

**Zielform:**

<div class="alert alert-block alert-warning">
    
* Ein Punkt in einer ungeordneten Liste
* Ein weiterer Punkt in einer ungeordneten Liste
    * Ein Unterpunkt, um vier Leerzeichen eingerückt
* Statt * funktionieren auch + oder -

</div> 

### Nummerierte-Listen

<div class="alert alert-block alert-success">

**Ausgangsform:**

```Markdown
1. Ein Punkt in einer geordneten Liste
2. Ein weiterer Punkt; bei der Eingabe muss nicht auf irgendeine Reihenfolge geachtet werden, sondern nur darauf, dass es beliebige Ziffern sind
1. Noch ein Punkt, der zeigt, dass auch die mehrfache Angabe derselben Ziffer möglich ist
```
</div>

**Zielform:**

<div class="alert alert-block alert-warning">
    
1. Ein Punkt in einer geordneten Liste
2. Ein weiterer Punkt; bei der Eingabe muss nicht auf irgendeine Reihenfolge geachtet werden, sondern nur darauf, dass es beliebige Ziffern sind
1. Noch ein Punkt, der zeigt, dass auch die mehrfache Angabe derselben Ziffer möglich ist

</div>

## Tabellen einbinden und Inhalte in den Zellen ausrichten

Das direkte Erstellen von Tabellen aus Daten mittel ***R*** ist der einfachste Weg - manchmal muss man aber auch Tabellen manuell erstellen. Hier hält Markdown folgenden Syntax bereit:

<div class="alert alert-block alert-success">

**Ausgangsform:**

```Markdown
Id| Syntax     | Description 
--|:----------:|:-----------:
1 |Header      | Something  here
2 |More here   | Text
    
```
</div>

**Zielform:**

<div class="alert alert-block alert-warning">
    
Id| Syntax    | Description 
--|:---------:|:-----------:
1 |Header     | Something  here
2 |More here  | Text

</div> 

Um die Ausrichtung in den verschiedenen Spalten zu bestimmen, muss man die Doppelpunkte entsprechend setzen.

**Ausgangsform:**

```Markdown
| Spalte 1 | Spalte 2 | Spalte 3 |
| :------: | :------- | -------: |
| zentriert| links    | rechts   |
    
```
</div>

**Zielform:**

<div class="alert alert-block alert-warning">
    
| Spalte 1 | Spalte 2 | Spalte 3 |
| :------: | :------- | -------: |
| zentriert| links    | rechts   |

</div> 

# Extras und Tipps

Um Boxen um einen Text zu legen, kann man folgendes in Jupyter Notebooks nutzen:

```HTML
<div class="alert alert-block alert-info">
<b>Tip:</b> Use blue boxes (alert-info) for tips and notes. 
If it’s a note, you don’t have to include the word “Note”.
</div>

<div class="alert alert-block alert-warning">
<b>Example:</b> Use yellow boxes for examples that are not 
inside code cells, or use for mathematical formulas if needed.
</div>

<div class="alert alert-block alert-success">
<b>Up to you:</b> Use green boxes sparingly, and only for some specific 
purpose that the other boxes can't cover. For example, if you have a lot 
of related content to link to, maybe you decide to use green boxes for 
related links from each section of a notebook.
</div>

<div class="alert alert-block alert-danger">
<b>Just don't:</b> In general, avoid the red boxes. These should only be
used for actions that might cause data loss or another major issue.
</div>

<code style="background:yellow;color:black">Useful for highlighting to grab the attention of the reader towards certain points.</code>
```
